# **Importing the libraries:**

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import sklearn
import scipy
import seaborn as sns
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

# **Loading the data:**

```python
data = pd.read_csv("/content/drive/MyDrive/Regression/Simple linear regression.csv")
```

```python
data
```

```python
data.describe()
```

# **Create our first regression:**

# **Creating dependent and independent variable:**

```python
y = data["GPA"].values.reshape(-1,1)
x = data["SAT"].values.reshape(-1,1)
```

# **Creating the scatter plot for this data:**

```python
plt.scatter(x,y,marker = "*",color = "r",s = 180)
plt.xlabel("Scholastic Apptitude Test",fontsize = 16,color = "g")
plt.ylabel("Grade point average",fontsize = 16,color = "g")
plt.title("Scatter plot",fontsize = 16,color = "g")
plt.show()
```

# **Creating with the help of simple linear regression :**

```python
x1 = sm.add_constant(x)
result = sm.OLS(y,x1).fit()
result.summary()
```

# **Train the regression model:**

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 1/2,random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

# **Visualize the regression result with best fit line:**

```python
plt.scatter(x_train,y_train,marker = "o",color = "b",s = 130)
plt.plot(x_train,regressor.predict(x_train),color = "y")
plt.xlabel("Scholastic Apptitude Test",fontsize = 16,color = "g")
plt.ylabel("Grade point average",fontsize = 16,color = "g")
plt.title("Scatter plot",fontsize = 16,color = "g")
plt.show()
```
